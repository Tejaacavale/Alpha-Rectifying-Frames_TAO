This notebook shows a idealised toy example to ilustrate the idea behind Monte Carlo Bias Estimation.

```python
import numpy as np
import mcbe
import matplotlib.pyplot as plt
import os
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"
```

Examples: Icosahedron and random frame on the sphere

```python
phi = (1+np.sqrt(5))/2
ico = np.array([[0,1,phi],[0,1,-phi],[0,-1,phi],[0,-1,-phi],
                [1,phi,0],[-1,phi,0],[1,-phi,0],[-1,-phi,0],
                [phi,0,1],[phi,0,-1],[-phi,0,1],[-phi,0,-1]])/(np.sqrt(1+phi**2))
d = 3
num_points = 12
ran = mcbe.norm_row(np.random.randn(num_points,d))[0]
```

We use the fact that the expected value of the covering radius of a d dimensional sphere $ \mathbb{E}( \rho (X_n)) âˆ¼ B_d ((log N )/N)^(1/d)$ as $ N -> \inf $ to get a feeling about the number of sampled points and the covering radius of the unit sphere.

```python
# fix N and get the covering radius 
N = 2000 
eps = mcbe.solve_eps(d,N)
print("asymptotically expected covering radius for", N, "points:",eps)
```

```python
# fix covering radius and get N 
eps = 0.25
N = mcbe.solve_N(d, eps)
print("number of sampling point to get a asymptotically expected covering radius", eps, ":", N)
```

To insure (expected?) injectivity on the whole sphere one can later subtract the covering radius from the estimated epsilon.

We compute $\alpha$ using Monte Carlo Bias Estimation with N random approximation points on the sphere and visualize the percent of injective samples in a random test set per iteration.

on the random frame:

```python
alpha_ran, subframes_ran, points_ran = mcbe.mcbe(ran, N, distribution="sphere", give_subframes = True, plot=True,iter_plot=500)
print('alpha for a random frame:', alpha_ran)
```

on the icosaeder:

```python
alpha_ico, subframes_ico, points_ico = mcbe.mcbe(ico, N, distribution="sphere", give_subframes=True, plot=True,iter_plot=5000)
print('alpha for the Icosahedron frame:', alpha_ico)
```

We know that the maximum bias on the icosaeder is $ \phi / (1 + \phi^2) $, so we can do a error analysis.

```python
print('Approximation error on the icosaeder: ', np.linalg.norm(alpha_ico - phi/(1+phi**2)))
```

To ensure (expected?) injectivity on the whole sphere we can now subtract the covering radius epsilon from the expected alpha:

```python
print('alpha for a random frame:', alpha_ran)
```

```python
print('alpha minus epsilon for a random frame:', alpha_ran - eps)
```

```python
print('alpha for the Icosahedron frame:', alpha_ico)
```

```python
print('alpha minus epsilon for the Icosahedron frame:', alpha_ico - eps)
```

Since we know the maximum bias on the icosahedron frame we can do a error analysis again but we need to keep in mind that the approximated bias is now smaller than the real bias.

```python
print('Approximation error on the icosaeder: ', np.linalg.norm(alpha_ico - eps -phi/(1+phi**2)))
```
